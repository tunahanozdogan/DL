```python
import matplotlib.pyplot as plt
from matplotlib.backends.backend_agg import FigureCanvasAgg as FigureCanvas
import numpy as np
from PIL import Image, ImageDraw, ImageFont

# Kanvası oluştur
width, height = 500, 500
canvas = Image.new('RGB', (width, height), (255, 255, 255))

# ImageDraw modülü ile kanvas üzerinde çizim yap
draw = ImageDraw.Draw(canvas)

# Fontu belirle
font = ImageFont.truetype("arial.ttf", 100)

# "SELİM" yazısını kanvasa çiz
draw.text((100, 200), "SELİM", fill=(0, 0, 0), font=font)

# Kanvası numpy array'e dönüştür
canvas_array = np.array(canvas)

# Numpy array'ini matplotlib grafiğine dönüştür
fig, ax = plt.subplots()
ax.imshow(canvas_array)

plt.show()
```
